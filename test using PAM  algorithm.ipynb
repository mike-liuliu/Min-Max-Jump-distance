```python
from mmj_functions import *
import kmedoids
 
```

```python
test_data_145 = pickle.load(  open( "./data/test_data_145.p", "rb" ) ) 
true_K_145 =  pickle.load(  open( "./data/true_num_of_145.p", "rb" ) ) 
```

```python
data_id  = 69

X = test_data_145[data_id] 
num_clusters =  true_K_145[data_id]

mmj_distance_matrix = pickle.load( open( f"./mmj_distance_matrix_precomputed/mmj_r_data_{data_id}.p", "rb" ) ) 
 
# Fit model
km_model = kmedoids.fasterpam(mmj_distance_matrix, num_clusters)


label = km_model.labels

plot_2D_or_3D_data(X, label)
```
